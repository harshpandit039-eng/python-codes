```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Customer Zipcode'].fillna(df['Customer Zipcode'].mean(),inplace = True)
```

```python
df.isnull().sum()
```

```python
df['Additional Order items'].fillna(df['Additional Order items'].mode()[0],inplace = True)
```

```python
df.isnull().sum()
```

```python
x=df['Category Name']
y=df['Sales']
plt.bar(x,y,color='g')
plt.xlabel('Category Name')
plt.ylabel('Sales')
plt.title('Sales by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
filtered_df=df[df['Order Date'].dt.year==2017]
Daily_sales_filtered = filtered_df.groupby('Order Date')['Sales'].sum().reset_index()
Daily_sales_filtered=Daily_sales_filtered.sort_values('Order Date')
    
plt.figure(figsize=(14,6))
plt.plot(Daily_sales_filtered['Order Date'],Daily_sales_filtered['Sales'],color = 'darkblue',linewidth=2,marker='o',markersize=5)
plt.xlabel('order Date')
plt.ylabel('Sales')
plt.title('Sales Over Year')
plt.xticks(rotation = 45)
plt.grid(True,alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
df['Year']=df['Order Date'].dt.year
df['Months']=df['Order Date'].dt.month
df['Day']=df['Order Date'].dt.day
```
