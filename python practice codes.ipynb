```python
def count_special_char(string):
    sp_char = "!@#$%%^&*()_-+=|"
    count = 0
    for special in string:
        if special in sp_char:
            count+=1
    return count
message = "hi how are you @hope#you are fine&"
counting = count_special_char(message)
print(counting)
            
```

```python
# what are python decoraters
#python decoraters are way to modify the oe extend the behavior of the function without changing the actual code
def decorator(func):
    def wrapper():
        print("before hello function run")
        func()
        print("after hello fucntion runs")
    return wrapper
@decorator
def say_hello():
    print("hello")
say_hello()
```
